```python
#networkx library provides method to add nodes and create edges
#matplotlib creates figure, plots, lines, etc
import networkx as nx
import matplotlib.pyplot as plt

#variable Draw initializes the graph
Draw = nx.Graph()

#color of the nodes i.e. yellow
color_map = []
color_map.append('yellow')

#arranging the position of each nodes
Draw.add_node('Uni', Pos = (1,4));
Draw.add_node('Aung', Pos = (0,3));
Draw.add_node('MD', Pos = (2,3));
Draw.add_node('Nima', Pos = (1,2));
Draw.add_node('Dawa', Pos = (0,1));
Draw.add_node('Sherpa', Pos = (2,1));
Draw.add_node('Temba', Pos = (1,0));
Draw.add_node('Pema', Pos = (3,2));
Draw.add_node('Luffy', Pos = (4,2));
Draw.add_node('OppaHero', Pos = (5,2))

#connecting the appropriate nodes
Draw.add_edges_from([('Uni','Aung'),('Uni','Nima'),('Uni','MD'),
                    ('Aung','MD'),('Aung','Nima'),('Aung','Dawa'),
                    ('Dawa','Nima'),('Dawa','Sherpa'),('Dawa','Temba'),
                    ('MD','Nima'),('MD','Sherpa'),('MD','Pema'),
                    ('Sherpa','Nima'),('Sherpa','Temba'),('Sherpa','Pema'),
                    ('Temba','Nima'),
                    ('Luffy','Pema'),('Luffy','OppaHero')])

#attaching and positioning all the nodes with edges and displaying in the graphs
Pos = nx.get_node_attributes(Draw,'Pos')

#drawing the network with the given info if the info is true. 
nx.draw(Draw,Pos,node_color = color_map,with_labels = True, edge_color = 'purple')

plt.show()
```
